```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sci

mat = sci.loadmat('c2.mat')

bar_input = mat['bar_input']
ebars = mat['ebars']

x = np.linspace(0, 3000, 1500)
y = bar_input[:,0]
error = ebars[:,0]

ax = plt.subplot()
ax.plot(x, y,color='k',linewidth=0.8)

plt.fill_between(x, y-error, y+error,facecolor='0.5')
ax.set_ylabel('Angle')
ax.set_xlabel('Time(ms)')
plt.axis([0, 3000, -1, 8])
plt.rcParams.update({'font.size': 14})
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(0.5)
ax.spines['left'].set_linewidth(0.5)
plt.savefig('Chat_M2_Baseline_7Hz.png', dpi=200)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sci
from matplotlib.ticker import AutoMinorLocator

plt.rcParams['font.family'] = "Arial"
plt.rcParams.update({'font.size': 10})
plt.rcParams['savefig.bbox'] = 'tight'

mat = sci.loadmat('fft.mat')
minor_locator = AutoMinorLocator(2)
minor_locatorx = AutoMinorLocator(2)

bar_input = mat['bar_input']
ebars = mat['ebars']

x = np.linspace(0, 250, 226) #Frequency spacing 
y1 = bar_input[0:50,0]
error1 = ebars[0:50,0]
y2 = bar_input[0:50,6]
error2 = ebars[0:50,6]
x = x[0:50]

ax = plt.subplot()
ax.plot(x, y1,'k', label = 'Baseline')
ax.plot(x,y2,'r', label = '+7 Days')
plt.fill_between(x, y1-error1, y1+error1,facecolor='0.5')
plt.fill_between(x, y2-error2, y2+error2,facecolor='0.5')
ax.set_ylabel('Amplitude')
ax.set_xlabel('Frequency')
plt.axis([-4, 50, -0.01, 0.15])
plt.yticks([0,0.05,0.1,0.15])
plt.xticks([0,10,20,30,40,50])
ax.yaxis.set_minor_locator(minor_locator)
ax.xaxis.set_minor_locator(minor_locatorx)
plt.tick_params(which='major', length=5)
plt.tick_params(which='minor', length=3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
fig = plt.gcf()
fig.set_size_inches(2,2)
ax.legend()

fig.savefig('FFT_spectra.svg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io as sci

plt.rcParams['font.family'] = "Arial"
plt.rcParams.update({'font.size': 10})
plt.rcParams['savefig.bbox'] = 'tight'

mat = sci.loadmat('Hex.mat')

bar_input = mat['Data']
control = bar_input[:,0]
hexa = bar_input[:,1]
nifed = bar_input[:,2]
pre = bar_input[:,3]

x = np.linspace(0, 250, 226) #Frequency spacing 

ax = plt.subplot()
ax.plot(x[1:90], control[1:90],'k', label = '+5 Control')
ax.plot(x[1:90], hexa[1:90],'r', label = '+5 Hexamethonium')
ax.plot(x[1:90], nifed[1:90],'b', label = '+5 Nifedipine')
ax.plot(x[1:90], pre[1:90],'0.4', label = 'Pre-Lesion')
ax.set_ylabel('Amplitude')
ax.set_xlabel('Frequency')
plt.axis([-4, 70, -0.01, 0.25])
plt.yticks([0,0.05,0.1,0.15,0.2,0.25])
plt.xticks([0,10,20,30,40,50])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
fig = plt.gcf()
#fig.set_size_inches(4,4)
ax.legend()

fig.savefig('Hex_spectra.svg')
plt.show()
```

```python
cm = np.mean(control[5:8])
hm = np.mean(hexa[5:8])
nm = np.mean(nifed[5:8])

fig, ax = plt.subplots()
width = 0.45

ind = 3

plt.bar(ind, cm, width, color='0.9', capsize=5, label = 'Control')
plt.bar(ind+width, hm, width, color='0.6', capsize=5, label = 'Hexamethonium')
plt.bar(ind+2*width, nm, width, color='0.3', capsize=5, label = 'Nifedipine')

# add some text for labels, title and axes ticks
plt.ylabel('Average Power 5-8Hz')
ax.set_ylim(0,0.25)
ax.set_xlim(2,5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
ax.tick_params(axis = 'x', bottom = False)
ax.set_xticklabels({})
ax.legend()
#fig.set_size_inches(4,4)
ax.set_title('+5 days after lesion')

fig.savefig('Hex_bar.png')
plt.show()
```
